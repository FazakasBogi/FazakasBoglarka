```python
# MULTI-ANSWER HELPER - split listed answers and map each answer separately
import re
import unicodedata


def harmonization_norm(value):
    value = "" if pd.isna(value) else str(value)
    value = value.strip().lower()
    value = "".join(
        char for char in unicodedata.normalize("NFKD", value)
        if not unicodedata.combining(char)
    )
    value = re.sub(r"\s+", " ", value)
    return value


def unique_keep_order(values):
    seen = set()
    out = []
    for value in values:
        if value not in seen:
            seen.add(value)
            out.append(value)
    return out


def harmonize_multi_answer_cell(value, answer_map):
    if pd.isna(value):
        return value

    text = str(value).strip()
    if text == "":
        return value

    normalized_map = {
        harmonization_norm(key): mapped
        for key, mapped in answer_map.items()
    }

    # If the source export already used semicolons, harmonize each selected option separately.
    if ";" in text:
        parts = [part.strip() for part in text.split(";") if part.strip()]
        mapped_parts = [harmonize_multi_answer_cell(part, answer_map) for part in parts]
        return "; ".join(unique_keep_order(mapped_parts))

    text_norm = harmonization_norm(text)

    # Exact match for one selected option.
    if text_norm in normalized_map:
        return normalized_map[text_norm]

    # Some EN/RO exports store multiple selected options with commas. Detect known options inside the text.
    matches = []
    for key_norm, mapped_value in sorted(normalized_map.items(), key=lambda item: len(item[0]), reverse=True):
        if len(key_norm) >= 5 and key_norm in text_norm:
            matches.append(mapped_value)

    if matches:
        return "; ".join(unique_keep_order(matches))

    return value


def harmonize_multi_answer_column(column_name, answer_map):
    master_df[column_name] = master_df[column_name].map(
        lambda value: harmonize_multi_answer_cell(value, answer_map)
    )
```
